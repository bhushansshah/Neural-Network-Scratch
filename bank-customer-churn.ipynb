## Loading the dataset

```python
dataset = pd.read_csv('/kaggle/input/bank-customer-churn-datasetcsv/Churn Modeling.csv')
```

```python
dataset.describe()
```

## Shuffling a dataset

```python
dataset = dataset.sample(frac=1)
DATASET_SIZE = len(dataset)
TRAIN_SIZE = int(0.8 * DATASET_SIZE)
TEST_SIZE = DATASET_SIZE - TRAIN_SIZE
train_dataset = dataset.head(TRAIN_SIZE)
test_dataset = dataset.tail(TEST_SIZE)
```

```python
print('Train size - ', TRAIN_SIZE)
print('Test size - ', TEST_SIZE)
```

## Removing the unwanted columns

```python
train_dataset = train_dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
test_dataset = test_dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

## Reseting the index

```python
train_dataset.reset_index(drop=True, inplace=True)
test_dataset.reset_index(drop=True, inplace=True)
```

```python
train_dataset
```

# One hot encoding of geography and gender columns

## Finding the unique values in geography and gender

```python
geography_unique = dataset['Geography'].unique()
```

```python
gender_unique = dataset['Gender'].unique()
```

```python
print('Unique values in geography - ', geography_unique)
print('Unique values in gender - ', gender_unique)
```

## One hot encoding

```python
for location in geography_unique:
    col = (train_dataset['Geography'] == location).astype(int)
    train_dataset[location] = col
    col = (test_dataset['Geography'] == location).astype(int)
    test_dataset[location] = col
for gender in gender_unique:
    col = (train_dataset['Gender'] == gender).astype(int)
    train_dataset[gender] = col
    col = (train_dataset['Gender'] == gender).astype(int)
    test_dataset[gender] = col
```

## Removing the geography and gender columns

```python
train_dataset = train_dataset.drop(['Geography', 'Gender'], axis=1)
test_dataset = test_dataset.drop(['Geography', 'Gender'], axis=1)
```

```python
train_dataset
```

## Standardizing few columns

```python
cols = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']
training_means = train_dataset[cols].mean(axis=0).to_numpy()
training_std = train_dataset[cols].std(axis=0).to_numpy()
```

```python
train_dataset[cols] = (train_dataset[cols] - training_means) / training_std
test_dataset[cols] = (test_dataset[cols] - training_means) / training_std
```

```python
train_dataset
```

```python
train_dataset.to_csv('preprocessed_training_data.csv', index=False, header=True)
test_dataset.to_csv('preprocessed_testing_data.csv', index=False, header=True)
```
